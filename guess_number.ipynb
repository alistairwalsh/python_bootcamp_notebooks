```python
import random

game_on = True

answer = random.randint(1,100)

print('Pick a number between 1 and 100')

while game_on:
    
    
    guess = int(input('What is your guess? '))

    if guess == answer:
        print('Congratulations!!!!')
        game_on = False
    
    elif guess > answer:
        print('The number is lower')
    
    else:
        print('The number is higher')
```
